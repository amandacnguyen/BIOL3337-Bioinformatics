```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv('SC3seq_Cy_ProcessedData.txt', sep='\t')
df.head()
```

```python
# Drop the first column of df
df = df.iloc[:, 1:] # select all rows, and all columns except the first one.

# Set the macFas5_gene_symbol column to be the row names
df.index = df['macFas5_gene_symbol']
# Drop the macFas5_gene_symbol column
df = df.drop('macFas5_gene_symbol', axis=1)

df.head()
```

```python
# log transform values in df
df = df.apply(lambda x: np.log2(x + 1))
df.head()
```

```python
# Select all genes at E06 and E09

gp1_all = df.loc[:, df.columns.str.contains('E06')]
gp2_all = df.loc[:, df.columns.str.contains('E09')]
print(gp1_all.shape, gp2_all.shape)
```

```python
# Perform t-test for all genes

t_res = stats.ttest_ind(gp1_all.T, gp2_all.T)
```

```python
# P-values for all genes
t_res.pvalue
```

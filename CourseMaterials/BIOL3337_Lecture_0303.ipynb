```python
nanog_e06 = df.loc['NANOG', df.columns.str.contains('E06')]

# plot a histogram for nanog_e06
nanog_e06.hist()
print(nanog_e06.mean())

# plot a vertical line indicating the mean
plt.axvline(nanog_e06.mean(), color='r')
```

```python
observed_mean = nanog_e06.mean()

# Randomly draw 43 elements from nanog_e06 with replacement
mean_random = nanog_e06.sample(43, replace=True).mean()
print(observed_mean, mean_random)

# Create an array with 1000 0s
means_random = np.zeros(1000)

# Loop 1000 times
for i in range(1000):
  mean_random = nanog_e06.sample(43, replace=True).mean()
  means_random[i] = mean_random

print(mean_random)
```

```python
# Plot the sampling distribution of the mean
plt.hist(means_random)

# Standard error of the mean:
sem = nanog_e06.sem()
print("Formula", sem)

# Standard error of the mean from simulation:
print("Simulation", means_random.std())
```

```python
gp1 = df.loc['CDX2', df.columns.str.contains('E06')]
gp2 = df.loc['CDX2', df.columns.str.contains('E09')]
print(gp1.shape, gp2.shape)
print(gp1.mean(), gp2.mean())
```

```python
plt.hist(gp1, label='E06', alpha=0.5)
plt.hist(gp2, label='E09', alpha=0.5)
plt.xlabel('log2(RPKM)')
plt.ylabel('Frequency')
plt.legend()
```

```python
# Perform a t test for gp1 and gp2
res = stats.ttest_ind(gp1, gp2)
print(res.pvalue)
```

```python
combined_data = pd.concat([gp1, gp2])
combined_data.shape
```

```python
observed_diff = gp2.mean() - gp1.mean()
print(observed_diff)

diffs = np.zeros(1000) # Is going be the sampling distribution of the difference
for i in range(1000):
  # Randomly choose 43 data points from combined_data
  sim_gp1 = combined_data.sample(43, replace=True)
  sim_gp2 = combined_data.sample(20, replace=True)
  sim_diff = sim_gp1.mean() - sim_gp2.mean()
  #print(sim_diff)
  diffs[i] = sim_diff

plt.hist(diffs)
```

```python
# p-values:
(diffs > observed_diff).sum() / 1000
```

```python
# log2FC > 1 or log2FC < -1
# p < 0.05
# It is a differentially expressed gene

print(observed_diff)
```

```python
# Perform t-test for all genes
gp1_all = df.loc[:, df.columns.str.contains('E06')]
gp2_all = df.loc[:, df.columns.str.contains('E09')]
```

```python
print(gp1_all.shape, gp2_all.shape)
```

```python
gp1_all.head()
```

```python
# ttest_all = stats.ttest_ind(gp1_all, gp2_all)
# Fails bc performs column by column; requires axis change

ttest_all = stats.ttest_ind(gp1_all, gp2_all, axis = 1)
```

```python
print(ttest_all.pvalue)
```

```python
print(ttest_all.pvalue[0])
print(ttest_all.pvalue[-1])
```

```python
print(ttest_all.pvalue.shape)
```

```python
print(gp1_all.iloc[-1])
```

```python
print(gp2_all.iloc[-1])
```

```python
# The number of differentially-expressed genes (DEGs) based on a p-value threshold
print((ttest_all.pvalue < 0.05).sum())
```

```python
# LogFC comparing E09 to E06
logFC = gp2_all.mean(axis=1) - gp1_all.mean(axis=1)  # Use axis = 1 for 2D
```

```python
# Number of DEGs based on a threhold of logFC
# logFC: log(expr_A) - log(expr_B) = log(expr_A/expr_B)
(logFC < -1).sum()
```

```python
plt.scatter(logFC, -np.log10(ttest_all.pvalue))
plt.xlabel('log2(FC)')
plt.ylabel('-log10(pvalue)')
plt.axhline(-np.log10(0.05), color='r')
plt.axvline(1, color='r')
plt.axvline(-1, color='r')
```

```python
pvalues = np.nan_to_num(ttest_all.pvalue, nan=1)
fdrs = stats.false_discovery_control(pvalues, method='bh')
```

```python
# FDR threshold to determine DEG
print((fdrs < 0.05).sum())
```
